```python
## Importing Libraries

import pywhatkit

from datetime import date
import os

import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
## Environment variables

os.environ['Group_Invite_Link'] = 'https://chat.whatsapp.com/CsAe0NnmW1n1eRf7P7IE61'
hour = os.environ['Hour'] = '16'
minute = os.environ['Minute'] = '02'

## Future task to import it from .env file
```

```python
class WhatsApp:
    
    def sendMessage(msg,group = os.environ['Group_Invite_Link']
                    ,hour = int(os.environ['Hour'])
                    ,minute = int(os.environ['Minute'])):

        groupId = group.split('/')[-1]
        # Using Exception Handling to avoid unprecedented errors

        try:

            # Sending message to receiver using pywhatkit
            pywhatkit.sendwhatmsg_to_group(groupId,msg,hour,minute)

        except:

            # Handling exception and printing error message

            f = open('error_log.txt',"a")

            f.write(today = date.today() +'\nAn Unexpected Error!\n\n')

    def verseUrl(m):

        ## Input as msg number
        ## O/p as url

        df = pd.read_csv('ChapVerse.csv')

        c = 0
        for i in df['Number']:
            for j in range(df.iloc[int(i)-1,2]):
                chapterNos, verseNos = i,j+1
                base = 'https://www.holy-bhagavad-gita.org/chapter/'  
                base = base + str(chapterNos)
                base = base + '/verse/' 
                base = base + str(verseNos)  
                base = base + '/'

                c = c + 1

                if c == m:
                    flag = True
                    return base

        return 'Invalid msg number'

    def findMsgNos():

        # Read the msg and figure out how many sent already

        f = open("PyWhatKit_DB.txt", "r")
        lines = f.readlines()

        c = 1
        for x in range(0,len(lines),5):
            c = c + 1
        return c
    
    def getMsg():

        # Using msg no, find verse

        # Getting url by passing msg sequence number
        URL = verseUrl(findMsgNos())

        # Getting page of specific url
        page = requests.get(URL)

        # Using soup for better reading
        soup = BeautifulSoup(page.content, "html.parser")

        # Finding the title
        title = str(soup.title)

        # Finding the verse
        soup.find(id='transliteration_wo_dia')

        # Cleaning to obtain message
        m2 = title.split(', The')[0]
        m3 = m2.replace('<title>','')

        verse = soup.find(id='transliteration_wo_dia').p
        t = str(verse)
        t2 = t.replace('<br/>','\n')
        t3 = t2.replace('<p>','').replace('</p>','')

        msg = m3 + '\n' + t3

        return msg
```
